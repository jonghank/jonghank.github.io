```julia
using HTTP, JSON, PyPlot

resp = HTTP.get("https://jonghank.github.io/ee787/files/fitting_outliers.json")
str = String(resp.body)
jobj = JSON.Parser.parse(str)

U = jobj["U"]["data"]
v = jobj["v"]["data"]

figure(figsize=(6,6), dpi=100)
plot(U,v, "o", alpha=0.5)
xlabel("U")
ylabel("v")
grid()
show()
```
