```julia
# load diabetes.data

using CSV

Diabete_Data = CSV.read("diabetes.data"; datarow = 2, delim='\t',types=fill(Float64,11));

n,d = size(Diabete_Data);
d = d-1;

Diabete_Data
```

```julia
# compute the optimal theta

X = zeros(n,d);
y = zeros(n,1);

for i=1:d
    X[:,i] = Diabete_Data[:,i];
end
X = [ones(n,1) X];
y = Diabete_Data[:,end];

theta_opt = X\y
```

```julia
# achieved loss

MSE = norm(X*theta_opt-y,2)^2/n
```

```julia
# plot

using PyPlot;

figure();
plot(y,y,"k");
plot(y,X*theta_opt,"o",alpha=0.4);
xlabel("y");
ylabel("predicted y")
axis("square");
grid("on");
```
